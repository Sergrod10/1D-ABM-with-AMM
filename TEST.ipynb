TEST OVERALL

```python
RANGE = list(range(1, 7 + 1, 2))
WINDOW = 5
SIZE = 10
FUNCS = [
    ('price', lambda info, w: info.prices)
]

traders = list()
before = list()
after = list()
configs = [[1, 5, 7, 5], [1, 7, 1, 5], [3, 1, 1, 3], [3, 1, 3, 1], [3, 1, 7, 3]]

for n_rand, n_fund, n_chart, n_univ in configs:
    for is_mm in range(2):
        exchange = ExchangeAgent(volume=1000)
        amm = UniswapV2()
        simulator = Simulator(**{
            'exchange': exchange,
            'amm': amm,
            'traders': [
                *[Random(exchange, 10 ** 3) for _ in range(n_rand)],
                *[Fundamentalist(exchange, 10 ** 3) for _ in range(n_fund)],
                *[Chartist(exchange, 10 ** 3) for _ in range(n_chart)],
                *[Universalist(exchange, 10 ** 3) for _ in range(n_univ)],
                *[MarketMaker(exchange, 10 ** 3) for _ in range(is_mm)]
            ],
            'arbitrage_traders': [*[Arbitrage(exchange, 10**3, amm) for _ in range(5)]],
        })
        info = simulator.info
        simulator.simulate(1500, silent=True)

        traders.append({'Random': n_rand, 'Fundamentalist': n_fund, 'Chartist': n_chart, 'Universalist': n_univ,
                        'MarketMaker': is_mm})
        print("CONFIGURATION IS: ", n_rand, n_fund, n_chart, n_univ)
        plot_price(simulator.info)
        plot_price_amm(simulator.amm_info)
        plot_cash_amm(simulator.amm_info)
        plot_assets_amm(simulator.amm_info)
        plot_equities_amm(simulator.amm_info)
        print("Participated: ")
        print(simulator.amm_info.participated[-1])
        print("-------------------------------------------------")
        print("Queue: ")
        print(simulator.amm_info.arbs_queue[150:160])
```
