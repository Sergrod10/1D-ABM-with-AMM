```python
import optuna
from optuna.trial import TrialState

STORAGE = "sqlite:///calibration_mae_2(10000).db"
STUDY_NAME = "calibration_mae_2(10000)"

study = optuna.load_study(study_name=STUDY_NAME, storage=STORAGE)

# Для single-objective
direction = study.direction  # StudyDirection.MINIMIZE / MAXIMIZE

def is_better(new, best):
    if best is None:
        return True
    if direction.name == "MINIMIZE":
        return new < best
    return new > best

best_value = None
best_number = None

for t in sorted(study.trials, key=lambda x: x.number):
    if t.state == TrialState.COMPLETE:
        if is_better(t.value, best_value):
            best_value = t.value
            best_number = t.number

        print(
            f"Trial {t.number} finished with value: {t.value} "
            f"and parameters: {t.params}. "
            f"Best is trial {best_number} with value: {best_value}."
        )

    elif t.state == TrialState.FAIL:
        reason = t.system_attrs.get("fail_reason", "No fail_reason in system_attrs.")
        print(
            f"Trial {t.number} failed with parameters: {t.params}. "
            f"Reason: {reason}"
        )

    elif t.state == TrialState.PRUNED:
        print(f"Trial {t.number} pruned. params: {t.params}")
```
