```python
# тест с разными std
N_RUNS = 5
METRIC_KEYS = [
    'mean_on_ret2',
    'std_on_ret2',
    'min_on_ret2',
    'max_on_ret2',
    'skewness_on_ret',
    'kurtosis_on_ret2',
]

EXEC_COMPARISON_DF = pd.DataFrame()

def exec(n_rand, n_fund, n_chart, n_univ, n_mm, stdd: float = 25):
    global EXEC_COMPARISON_DF
    run_rows = []

    for run_idx in range(N_RUNS):
        random.seed(42 + run_idx)
        np.random.seed(42 + run_idx)

        exchange = ExchangeAgent(volume=10000, std=stdd)
        simulator = Simulator(**{
            'exchange': exchange,
            'traders': [
                *[Random(exchange, 10**4) for _ in range(n_rand)],
                *[Fundamentalist(exchange, 10**4) for _ in range(n_fund)],
                *[Chartist(exchange, 10**4) for _ in range(n_chart)],
                *[Universalist(exchange, 10**4) for _ in range(n_univ)],
                *[MarketMaker(exchange, 10**4) for _ in range(n_mm)],
            ],
        })

        simulator.simulate(1000, silent=True)
        prices2 = np.array(simulator.info.prices)
        model_params = pipeline(prices2, is_print=0)

        run_row = {'run_idx': run_idx, 'stdd': stdd}
        for key in METRIC_KEYS:
            run_row[key] = model_params.get(key, np.nan)
        run_rows.append(run_row)

    run_metrics_df = pd.DataFrame(run_rows)
    mean_metrics = run_metrics_df[METRIC_KEYS].mean(numeric_only=True)

    summary_row = {
        'n_rand': n_rand,
        'n_fund': n_fund,
        'n_chart': n_chart,
        'n_univ': n_univ,
        'n_mm': n_mm,
        'stdd': stdd,
        **mean_metrics.to_dict(),
    }
    EXEC_COMPARISON_DF = pd.concat([EXEC_COMPARISON_DF, pd.DataFrame([summary_row])], ignore_index=True)

    print(f'Average metrics across {N_RUNS} runs (stdd={stdd}):')
    for key in METRIC_KEYS:
        print(f'{key}: {mean_metrics[key]}')

    return run_metrics_df, EXEC_COMPARISON_DF.copy()

_, comparison_df = exec(35, 79, 51, 22, 1, 0.1)
_, comparison_df = exec(35, 79, 51, 22, 1, 1)
_, comparison_df = exec(35, 79, 51, 22, 1, 10)
_, comparison_df = exec(35, 79, 51, 22, 1, 100)
_, comparison_df = exec(35, 79, 51, 22, 1, 1000)

comparison_df.sort_values('stdd').reset_index(drop=True)
```
